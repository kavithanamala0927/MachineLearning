## Seaborn Tutorial

### Distribution Plots

    1.distplot
    2.joinplot
    3.pairplot


```python
import seaborn as sns

data = sns.load_dataset("tips")

data.head()
```

### Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature selection


```python
print (data.dtypes)
data.corr()
```

```python
sns.heatmap(data.corr())
```

## JoinPlot


A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation 
It is usually a scatter plot, a hexbin plot, a 2D histogram or a 2D density plot

## Bivariate Analysis

```python
sns.jointplot(x='tip', y = 'total_bill', data=data, kind='hex')
```

```python
sns.jointplot(x='tip', y = 'total_bill', data=data, kind='reg')
```

## Pair plot

A Pairplot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: pair plots are just elaborations on this., showing all variables paired with all other variables.

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue='sex')
```

## Dist Plot

Dist plot helps us to check the distribution of the columns feature.

```python
sns.distplot(data['tip'])
```

```python
sns.distplot(data['tip'], kde=False, bins=10)
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points

  1.box plot

  2.violinplot
 

  3.countplot


  4.bar plot 

```python
##Count plot
sns.countplot('sex', data=data)
data['sex'].value_counts()
```

```python
sns.countplot('smoker', data=data)
```

```python
sns.countplot('day', data=data)
```

```python
sns.countplot(y='sex',data=data)
```

```python
##Bar plot
sns.barplot(x='total_bill', y='sex', data=data)
```

```python
sns.barplot(x='sex',y='total_bill', data=data)
```

#### Box Plot 

A box and whisker plot (sometimes called as boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('smoker','total_bill', data=data)
```

```python
sns.boxplot('sex', 'total_bill', data=data)
```

```python
sns.boxplot(x='day', y='total_bill', data=data, palette='rainbow')
```

```python
sns.boxplot(data=data, orient='v')
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=data)
```

### Violin Plot


Violin plot helps us to see both the distribution of the data in terms of Kernel density estimation and  the box plot 

```python
sns.violinplot(x='total_bill', y='day', data=data, palette='rainbow')
```
