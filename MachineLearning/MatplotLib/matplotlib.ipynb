## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension Numpy. It provides an object-oriented API for embedding plots into applications using generic-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+

Some of the major Pros of Matplotlib are:    
    

    . Generally easy to get started for simple plots
    . Support for custom labels and texts
    . Great control of every element in a figure
    . High-quality output in many formats
    . Very customizable in general

import matplotlib.pyplot as plt
%matplotlib inline

```python
import numpy as np 
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
##plotting using matplotlib
#plot scatter
plt.scatter(x,y,c='g')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Grapn in 2D")
plt.savefig("Test.png")
```

```python
##Plot
plt.plot(x,x*x,'r--')
```

```python
plt.plot(x,x*x,'r',linestyle='dashed', linewidth=3,markersize=50)
```

```python
##Creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,x*x,'g')

plt.subplot(2,2,3)
plt.plot(x,x+x,'y')

plt.subplot(2,2,4)
plt.plot(x,y*y,'b')
plt.subplot(2,2,5)
plt.plot(x,y,'l')
```

```python
#compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sine wave form")
plt.plot(x,y)
```

```python
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r')
```

```python
##Bar Plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
```

```python
#Histogram 
a = np.array([22,87,3,45,34,32,46,77,1,11,20])
plt.hist(a)
```

```python
##BoxPlot

data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```

```python
#PieChart
labels = 'Python', 'C++', 'Ruby', 'java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'lightskyblue', 'green', 'red']
explode = (0.1, 0, 0, 0) #explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
                                autopct = '%1.2f%%', shadow = True)
plt.axis('equal')
plt.show()

```
